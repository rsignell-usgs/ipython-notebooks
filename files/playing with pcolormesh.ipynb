```python
a=rand(3,4)
```

```python
x=arange(4)+0.5
y=arange(3)+0.5
[xx,yy]=meshgrid(x,y)
```

```python
pcolormesh(a,vmin=0,vmax=1)
plot(xx,yy,'ko');
```

```python
X=xx+1j*yy
X=X*exp(1j*pi/4)
xx2=X.real
yy2=X.imag
```

```python
pcolormesh(xx2,yy2,a,vmin=0,vmax=1,shading='flat')
plot(xx2,yy2,'ko');

```
