```python
# Display a remote JPEG URL in Matplotlib without saving to disk.  
# Thanks to Robert Kern for the help!
```

```python
from pylab import *
import urllib
import urllib2
import cStringIO
from PIL import Image
```

```python
# It's easy to load a remote image URL if it's a PNG:
url = 'http://optipng.sourceforge.net/pngtech/img/lena.png'
im = imread(urllib2.urlopen(url),format='png')
imshow(im)
```

```python
#  But here's how you have to do it if the URL is a JPG: 
#url = 'http://www.kitware.com/blog/files/128_1232259669.jpg'
url='http://4.bp.blogspot.com/-jUHXZ0jW4Lc/T1t3W0ex2BI/AAAAAAAAFeY/7TY0cTty1og/s1600/Fabio.jpg'
im = Image.open(cStringIO.StringIO(urllib.urlopen(url).read()))
imshow(im)
```

```python
# .... Because this doesn't work:
url = 'http://4.bp.blogspot.com/-jUHXZ0jW4Lc/T1t3W0ex2BI/AAAAAAAAFeY/7TY0cTty1og/s1600/Fabio.jpg'
im = imread(urllib2.urlopen(url),format='jpg')
imshow(im)
```
