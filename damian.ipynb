```python
import netCDF4
url='http://schumann.coastal.udel.edu:8080/thredds/dodsC/cb/test1'
nc = netCDF4.Dataset(url)
```

```python
var='PHYT1';itime=0;layer=0
lon = nc.variables['LON'][:,:]
lat = nc.variables['LAT'][:,:]
v = nc.variables[var][itime,layer,:,:]
times = nc.variables['TIME']
dtime = netCDF4.num2date(times[itime],units=times.units)
```

```python
fig=figure(figsize=(10,10))
ax=fig.add_subplot(111,aspect=1.0/cos(lat.mean() * pi / 180.0))
pcolormesh(lon,lat,v)
colorbar()
timestamp=dtime.strftime('%Y-%m-%d %H:%M:%S')
title('%s on layer %d: %s' % (var, layer, timestamp))
```
